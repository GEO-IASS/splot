# Oneshot mapping function

First, we read in the data.

````python
sou_js = gj.load(open('south.json'))
````

And, let's look at the columns that we're going to map.

````python
sou_js.features[0].properties.keys()[0:5]
````

We can map these attributes by calling them as commands to the oneshot choropleth mapping function. Initially we stated we were going for a grammar of graphics implementation. This would mean that you could add elements intelligently to the plot by overloading the python operators. So, think for a time series variable `HV90` with `HV` meaning housing values and `90` being the decade of the data,

```
oneshot.choropleth_map(data, 'FIPS', 'HV90', classification='Maximum Breaks') + oneshot.hoverplot(['HV90','HV80','HV70'], local=True)
```
would produce a Maximum Breaks map of housing values in the 1990s, with a mouseover (`hoverplot`) for each polygon, showing their housing value over the decades in the list. 

This proved to be too obnoxious, as the interface between Folium, the things that power Folium (Leaflet + D3), and Vincent were not suited to this kind of operator overloading. 

But, to show, here's an incredibly simple map command:


````python
oneshot.choropleth_map('./south.json', 'FIPS', 'HR90', save=True)
````

This produces a map using default classifications and color schemes and saves it to an html file. We set the function to have sane defaults, as per the spec we outlined. However, if the user wants to have more control, we have many options available. 

## There are arguments to change classification scheme:

````python
oneshot.choropleth_map('./south.json', 'FIPS', 'HR90', classification = 'Quantiles')
````

Most `PySAL` classifiers are supprorted. 

## Base Map Type

````python
oneshot.choropleth_map('./south.json', 'FIPS', 'HR90', classification = 'Jenks Caspall', tiles='Stamen Toner', save=True) 
````

We support the entire range of basemap types in Folium. 

## Color Scheme

````python
oneshot.choropleth_map('./south.json', 'FIPS', 'HR80', classification = 'Jenks Caspall', tiles='Stamen Toner', fill_color = 'PuBuGn', save=True) 
````

All color schemes are [Color Brewer](https://colorbrewer2.org) and simply pass through to `Folium` on execution. 

## Class numbers

````python
oneshot.choropleth_map('./south.json', 'FIPS', 'HR80', classification = 'Fisher-Jenks', classes=6, tiles='Stamen Toner', fill_color='PuBuGn',save=True) 
````

Folium supports up to 6 classes.

## Standardization

````python
oneshot.choropleth_map('./south.json', 'FIPS', 'HR90', classification = 'Quantiles', std='HR80' , tiles='Stamen Toner', fill_color='PuBuGn',save=True) 
````

The goal of the `std` argument was to implement a very flexible transformation argument that could take functions, fields, and static items. As of now, it only supports float/int scaling and division by another field, which would support the mapping of spatial rates or population rates.

# Conclusion

Overall, the function neatly encapsulates three components:

- plumbing large lists of arguments together to simplify mapping from JSON files
- searching for sane defaults (bounding box & zoom level)
- implementing classification and standardization useful for choropleth mapping

Due to some pretty strong limitations in `Folium` itself, its usefulness for rich, dynamic, and interactive visualizations built from small, atomic functions is probably limited. It would need to be extended quite significantly (or linked with even more packages with more fickle APIs) to usefully generate mixed-mode scientific visualizations. 

That said, this project delivers on its core two points: generate a function that converts `PySAL` objects into GeoJSON and a simple, short function that can map from GeoJSON quickly and easily. For simple and well-defined shapefiles, this can turn a shapefile into a map in two function calls:
